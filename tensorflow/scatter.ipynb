```python
import pandas as pd;
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python

url = 'https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv'
df = pd.read_csv(url)
df = df[df['year'] == 2007]
df.head()
```

```python
fig = sns.lmplot(x="lifeExp", y="gdpPercap", data=df, fit_reg=True, truncate=True, height=5, aspect=2.5)
fig.set_axis_labels("average age in country", "average GDP in country")
fig.set(title='Average age vs GDP in countries')
```
